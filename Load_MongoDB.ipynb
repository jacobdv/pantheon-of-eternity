```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import datetime as dt

# File to Load 
load_data = "StocksAndSentiment_basic.csv"

# Read from CSV
load_df = pd.read_csv(load_data)

load_df
```

```python
preprocess_df = load_df.loc[load_df['Year'] == 2017]
preprocess_df
```

```python
preprocess_df = preprocess_df.drop(['Unnamed: 0', 'Date'], axis=1)
preprocess_df = preprocess_df.reset_index()
preprocess_df = preprocess_df.drop(['index'], axis=1)
preprocess_df = preprocess_df.dropna()
preprocess_df
```

```python
data_binary_encoded = pd.get_dummies(preprocess_df)
data_binary_encoded = data_binary_encoded.dropna()
data_binary_encoded
```

```python
data_binary_encoded = data_binary_encoded.drop(['High','Low','Good_time_to_Buy','Uncertain','Bad_time_to_Buy', 'Relative'], axis=1)
data_binary_encoded
```
