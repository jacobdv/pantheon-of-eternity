```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import datetime as dt

# File to Load 
load_data = "StocksAndSentiment_basic.csv"

# Read from CSV
load_df = pd.read_csv(load_data)

load_df
```

```python
preprocess_df = load_df.loc[(load_df['Year'] == 2015) | (load_df['Year'] == 2016)]
preprocess_df
```

```python
preprocess_df = preprocess_df.drop(['Unnamed: 0', 'Year', 'Month', 'Day', 'Open', 'High', 'Low', 'Date', 'Good_time_to_Buy', 'Uncertain', 'Bad_time_to_Buy'], axis=1)
preprocess_df = preprocess_df.reset_index()
preprocess_df = preprocess_df.drop(['index'], axis=1)
preprocess_df = preprocess_df.dropna()
preprocess_df
```

```python
X_df = preprocess_df[['Symbol', 'Close', 'Volume']]

data_binary_encoded = pd.get_dummies(X_df)
data_binary_encoded = data_binary_encoded.dropna()
data_binary_encoded
```

```python
#One Hot Encoder Ticker Symbol
# Step 0: Reformat data
data = preprocess_df.values
#X = data[:, 0:3]
#x0 = data[:,0]
y = data[:, 0]
```

```python
from sklearn.preprocessing import LabelEncoder

# Step 1: Label-encode data set
label_encoder = LabelEncoder()
label_encoder.fit(y)
encoded_y = label_encoder.transform(y)
```

```python
for label, original_class in zip(encoded_y, y):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
from keras.utils import to_categorical

# Step 2: One-hot encoding
one_hot_y = to_categorical(encoded_y)
one_hot_y
```

```python
for index, row in preprocess_df.iterrows():
    preprocess_df.at[index,'Symbol'] = one_hot_y[index]
    #print(justy.loc[index,:])

preprocess_df
```

```python
xdata = data_binary_encoded.values
ydata = preprocess_df.values

X = xdata[:, 0:8]
y = ydata[:, 3]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
#np.asarray(X).astype(np.float32)
#len(X_train)
X_train = np.asarray(X_train).astype(np.float32)
X_train
```

```python
#len(y_train)
y_train = np.asarray(y_train).astype(np.float32)
y_train
```

```python
from tensorflow.keras.models import Sequential

model = Sequential()
```

```python
from tensorflow.keras.layers import Dense
number_inputs = 8
number_hidden_nodes = 4
model.add(Dense(units=number_hidden_nodes,
                activation='relu', input_dim=number_inputs))
```

```python
number_classes = 1
model.add(Dense(units=number_classes, activation='softmax'))
```

```python
model.summary()
```

```python
# Use categorical crossentropy for categorical data and mean squared error for regression
# Hint: your output layer in this example is using software for logistic regression (categorical)
# If your output layer activation was `linear` then you may want to use `mse` for loss
model.compile(optimizer='adam',
              loss='mean_squared_error',
              metrics=['accuracy'])
```

```python
# Fit (train) the model
model.fit(
    X_train,
    y_train,
    epochs=100,
    shuffle=True,
    verbose=2
)
```
