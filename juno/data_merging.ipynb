```python
import os
import pandas as pd
```

```python
car_stocks = ['RACE','F','GM','HMC','TTM','TM']
```

```python
completeDF = pd.DataFrame()
for stock in car_stocks:
    try:
        stockDF = pd.read_csv(f'../CSVs/{stock}.csv')
        stockDF['Symbol']  = stock
        combine = [stockDF, completeDF]
        completeDF = pd.concat(combine)
    except FileNotFoundError:
        print(f'No on {stock}.')
completeDF.drop(columns=['Unnamed: 0','OpenInt'], inplace=True)
```

```python
completeDF
```

```python
years = []
months = []
days = []

for index, row in completeDF.iterrows():
    print('Index:', index)
    print('Row:', row['Date'])
    print('-----')
    rowDate = row['Date'].split('-')
    rowYear = rowDate[0]
    rowMonth = rowDate[1]
    rowDay = rowDate[2]
    years.append(rowYear)
    months.append(rowMonth)
    days.append(rowDay)

completeDF['Year'] = years
completeDF['Month'] = months
completeDF['Day'] = days
```

```python
completeDF = completeDF[['Symbol','Year','Month','Day','Open','High','Low','Close','Volume']]
completeDF
```

```python
completeDF = completeDF.reset_index()
```

```python
completeDF.drop(columns=['index'], inplace=True)
```
