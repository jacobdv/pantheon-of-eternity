```python
# File to Load 
load_sentiment_detail = "UMICH-SOC38.csv"

# Read from CSV
sentiment_detail_df = pd.read_csv(load_sentiment_detail)

sentiment_detail_df.head(50)

#Relative: prices column (%LOW PRICES - %HIGH PRICES)
```

```python
# Caste Start_Time into datetime data type
#sentiment_detail_df["DateTime"] = sentiment_detail_df["Date"]
sentiment_detail_df = sentiment_detail_df.astype({'Date': 'datetime64'})
sentiment_detail_df
```

```python
sentiment_detail_df['Year'] = sentiment_detail_df['Date'].dt.year
sentiment_detail_df['Month'] = sentiment_detail_df['Date'].dt.month
sentiment_detail_df
```

```python
# rename columns
sentiment_detail_df.rename(columns={
    'Good Time<br>Prices are Low': "Good_Time_Prices_Low",
    'Good Time<br>Prices will increase': "Good_Time_Prices_will_increase",
    'Good Time<br>Interest rates low': "Good_Time_Interest_rates_low",
    'Good Time<br>Rising interest rates': "Good_Time_Rising_interest_rates",
    'Good Time<br>Times good': "Good_Time_Times_good",
    'Good Time<br>Fuel Efficiency': "Good_Time_Fuel_Efficiency",
    'Bad Time<br>Prices High': "Bad_Time_Prices_High",
    'Bad Time<br>Interest rates high': "Bad_Time_Interest_rates_high",
    'Bad Time<br>Can\'t Afford': "Bad_Time_Cannot_Afford",
    'Bad Time<br>Uncertain Future': "Bad_Time_Uncertain_Future",
    'Bad Time<br>Gas Prices': "Bad_Time_Gas_Prices",
    'Bad Time<br>Poor Selection': "Bad_Time_Poor_Selection",
    'Relative: prices': "Relative_prices",
    'Relative: rates': "Relative_rates"
}, inplace=True)

sentiment_detail_df
```

```python
# Caste Start_Time into datetime data type
#sentiment_detail_df["DateTime"] = sentiment_detail_df["Date"]
auto_sentiment_df = auto_sentiment_df.astype({'Date': 'datetime64'})
auto_sentiment_df
```

```python
auto_sentiment_df['Year'] = auto_sentiment_df['Date'].dt.year
auto_sentiment_df['Month'] = auto_sentiment_df['Date'].dt.month
auto_sentiment_df
```

```python
# rename columns
auto_sentiment_df.rename(columns={
    'Good time to Buy': "Good_time_to_Buy",
    'Bad time to Buy': "Bad_time_to_Buy"
    }, inplace=True)

auto_sentiment_df
```

```python
combined_basic_df = auto_stock_df.merge(auto_sentiment_df, on=['Month','Year'])
combined_basic_df
```

```python
combined_df = auto_stock_df.merge(sentiment_detail_df, on=['Month','Year'])
combined_df
```
